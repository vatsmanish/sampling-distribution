# Question5

### check if central limit theorm holds is in the variable monthly income

```python
mean=[]                              ## take it as a list
for i in range(1,10000):         ## A loop from 1 to 10000
    xi=np.random.choice(attrition['MonthlyIncome'],size=200)     ## randomly choice from monthly income in attrition dataset
    mean_x=np.mean(xi)                                          ## take mean
    mean.append(mean_x)
sns.distplot(mean,color='red')                ## plot a histogram
sns.distplot(xi,color='green')
plt.show()
```

#### it will give the mean and result of mean of population

```python
np.mean(xi)
```

#### it will give the standar deviation and standar deviatoin of mean

```python
np.std(mean)
```
